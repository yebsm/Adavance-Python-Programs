ATM PROGRAM WITH PIN

```python
wallet_amount = 0
activity_log = []
access_pin = 9876
counter = 0
permission = False

while counter < 3:
    print(f"Available chances :{3 - counter}")
    entered_value = int(input("Enter your pin :"))
    if entered_value == access_pin:
        permission = True
        break
    else:
        print("Wrong Pin")
        counter += 1

while permission:
    print("\n1 for Deposit")
    print("2 for withdraw")
    print("3 for Balance Inquiry")
    print("4 for All Transactions")
    print("5 for Exit\n")

    selection = int(input("Enter Your Choice :"))

    if selection == 1:
        value_add = int(input("Enter the amount for deposit :"))
        wallet_amount += value_add
        activity_log.append(f"Deposit of {value_add}")
        print(f"Deposit Successfull of {value_add}")

    elif selection == 2:
        value_sub = int(input("Enter the amount you want to withdraw :"))
        if value_sub > wallet_amount:
            print("Insuffecient Balance")
        else:
            wallet_amount -= value_sub
            activity_log.append(f"Withdraw of {value_sub}")
            print(f"Withdraw successfull of {value_sub}")

    elif selection == 3:
        print(f"Account balance :{wallet_amount}")

    elif selection == 4:
        if len(activity_log) == 0:
            print("No Transactions")
        else:
            for item in activity_log:
                print(item)

    elif selection == 5:
        print("Exiting...!")
        break

    else:
        print("Invalid Choice !!")

print("Thank you for using")
```

EVENT TRACKER SET

```python
approved_list = set()
denied_list = set()

attempts = int(input("Enter Entry attempt: "))

for _ in range(attempts):
    student_name = input("Enter student names: ").lower()
    formatted_name = "".join(student_name.split(" "))

    if formatted_name in approved_list:
        print(student_name, "already entered. entry rejected")
        denied_list.add(formatted_name)
    else:
        print(student_name, "Entry allowed")
        approved_list.add(formatted_name)

print("Event details:")
print("\nTotal unique entry :", len(approved_list))
for name in approved_list:
    print(name)

print("\nTotal rejected entry :", len(denied_list))
for name in denied_list:
    print(name)
```

ROOM ALLOTMENT USING DICTIONARY

```python
hotel_rooms = {201: None, 202: None, 203: None, 204: None, 205: None,
               206: None, 207: None, 208: None, 209: None, 210: None}

print("Available rooms :")
for room in hotel_rooms:
    print(room, end=" ")

attempt_count = int(input("Enter Entry attempt: "))

for _ in range(attempt_count):
    room_number = int(input("Enter the room number :"))
    if room_number in hotel_rooms and hotel_rooms[room_number] is not None:
        print("Room already alloted !!")
    elif room_number in hotel_rooms:
        guest_name = input("Enter your name :")
        hotel_rooms[room_number] = guest_name
        print(f"Room Alloted successfully to {guest_name}")

vacate_room = int(input("Enter the room to vacate :"))

if vacate_room not in hotel_rooms or hotel_rooms[vacate_room] is None:
    print("Cannot vacate Room or room already vacate")
else:
    hotel_rooms[vacate_room] = None
    print("Room checkout Successfully !!")

print("All Booked rooms details :")
for room in hotel_rooms:
    if hotel_rooms[room] is not None:
        print(f"{room} : {hotel_rooms[room]}")

print("All Vacate rooms details :")
for room in hotel_rooms:
    if hotel_rooms[room] is None:
        print(f"{room} : {hotel_rooms[room]}")

print("Thank You !!!")
```

ATM PROGRAM WITH FUNCTION

```python
def authenticate(user_input, master_key=9032):
    return user_input == master_key


def add_money(total, logs):
    amount = int(input("Enter the amount for deposit : "))
    total += amount
    logs.append(f"Deposit of {amount}")
    print(f"Deposit Successful of {amount}")
    return total


def remove_money(total, logs):
    amount = int(input("Enter the amount you want to withdraw : "))

    if amount > total:
        print("Insufficient Balance")
    else:
        total -= amount
        logs.append(f"Withdraw of {amount}")
        print(f"Withdraw Successful of {amount}")

    return total


def show_history(logs):
    if not logs:
        print("No Transactions")
    else:
        for record in logs:
            print(record)


def show_balance(total):
    print(f"Account balance : {total}")


current_amount = 0
activity_records = []
is_allowed = False

for attempt in range(3):
    print(f"Available chances : {3 - attempt}")
    entered_pin = int(input("Enter your pin : "))

    if authenticate(entered_pin):
        is_allowed = True
        break
    else:
        print("Wrong Pin")

while is_allowed:
    print("\n1 for Deposit")
    print("2 for Withdraw")
    print("3 for Balance Inquiry")
    print("4 for All Transactions")
    print("5 for Exit\n")

    choice = int(input("Enter Your Choice : "))

    if choice == 1:
        current_amount = add_money(current_amount, activity_records)

    elif choice == 2:
        current_amount = remove_money(current_amount, activity_records)

    elif choice == 3:
        show_balance(current_amount)

    elif choice == 4:
        show_history(activity_records)

    elif choice == 5:
        print("Exiting...!")
        break

    else:
        print("Invalid Choice !!")

print("Thank you for using")
```

LIBRARY MANAGEMENT

```python
library_data = {}
issued_records = {}

book_code = 100
total_books = int(input("Enter the no of books for entry :"))

for _ in range(total_books):
    title = input("Enter the name of the book :")
    writer = input("Enter the author of the book :")
    copies = int(input("Enter the no. of copies :"))

    library_data[book_code] = {
        "Code": book_code,
        "Title": title,
        "Writer": writer,
        "Stock": copies
    }

    print(f"{title} added successfully !!")
    book_code += 1

print("Books available :")
for key in library_data:
    print("ID", key, "Name :", library_data[key]["Title"],
          "No of copies available :", library_data[key]["Stock"])

issue_code = 500
issue_count = int(input("Enter the no. of book to allot :"))

for _ in range(issue_count):
    selected_id = int(input("Enter the book id to allot : "))

    if library_data[selected_id]["Stock"] > 0:
        member_name = input("Enter users name :")
        member_contact = int(input("Enter users phNo:"))

        issued_records[issue_code] = {
            "Book_ID": selected_id,
            "Member": member_name,
            "Book_Title": library_data[selected_id]["Title"],
            "Phone": member_contact
        }

        print(f"{selected_id} alloted to user successfully !!")
        library_data[selected_id]["Stock"] -= 1
        issue_code += 1
    else:
        print("Book not available")

print("Available books :")
for key in library_data:
    print("ID", key, "Name :", library_data[key]["Title"],
          "No of copies available :", library_data[key]["Stock"])

print("All Allotments")
for key in issued_records:
    print("Allot_ID", key, ": Book id :", issued_records[key]["Book_ID"],
          "Alloted to :", issued_records[key]["Member"])
```

TRAFFIC MANAGEMENT

```python
import time
color = "RED"
print("Signal Color :",color)
while True:
  col = input("Enter the color of the signal(RED,YELLOW,GREEN) or 0 to exit:")
  if col=="0":
    break
  col = col.upper()
  if col not in ("RED,YELLOW,GREEN") or col==color:
    print("Invalid Signal Color or Cannot be the same as previous color!!")
    continue
  tim = int(input("Enter the time in second :"))
  while(tim>0):
    print(f"Time left {tim} second")
    time.sleep(1)
    tim-=1
  color = col
  print("Color changed to :",color)
```
